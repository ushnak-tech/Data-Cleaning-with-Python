## seems like `block80` column is an address


```python
df['block80']=df['block80'].apply(lambda x:x.replace("(",'').replace(')',''))
```

```python
df['block80']=df['block80'].apply(lambda x:x.replace("BHK",'').replace('Bed','').replace(",",'').replace("1",'').replace("-",'').replace("2",''))
```

```python
df['block80']=df['block80'].apply(lambda x:" ".join(x.split()))
```

```python
def get_address(string):
  list_str=[i for i in string.split(' ')]
  list_str=list_str[-3:]
  string=''
  string=' '.join(list_str)
  return string
```

```python
df['block80']=df['block80'].apply(get_address)
```

```python
df['block80']=df['block80'].apply(lambda x:x.replace("in ",'').replace('by ','').replace("5",'').replace("4",'').replace("Balconies",'').replace("bedroom",''))
```

```python
df['block80']=df['block80'].apply(lambda x:" ".join(x.split()))
```

```python
import re
```

```python
addr={'JLT':'Jumehra Lake Tower','JBR':'Jumehra Beach Residence','JVC':'Jumehra Village Circle'}
def replace_func(string):
  for word, replacement in addr.items():
    string = re.sub(word, replacement, string)
  return string

```

```python
df['block80']=df['block80'].apply(replace_func)
```
